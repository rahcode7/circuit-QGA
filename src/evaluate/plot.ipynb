```python
import seaborn as sns
import pandas as pd 
import os 
from icecream import ic 
import matplotlib.pyplot as plt
```

```python
blip_df = pd.read_csv("models-BLIP-hf/results/summary_report.csv")
git_df = pd.read_csv("models-GIT-hf/results/base-384/summary_report.csv")

df = pd.concat([blip_df,git_df])


df = df[df['average'].isin(['micro'])]
ic(df)
```

```python

blip_base_df = pd.read_csv(" models-BLIP-hf/results/base-384/summary_report.csv")
desc_df = pd.read_csv(" models-BLIP-hf/results/desc/summary_report.csv")
ce_df = pd.read_csv(" models-BLIP-hf/results/wce/summary_report.csv")
size_df = pd.read_csv(" models-BLIP-hf/results/size-576/summary_report.csv")
size_768_df = pd.read_csv(" models-BLIP-hf/results/size-768/summary_report.csv")
ocr_desc_df = pd.read_csv(" models-BLIP-hf/results/ocr-desc/summary_report.csv")
ocr_df = pd.read_csv(" models-BLIP-hf/results/ocr/summary_report.csv")
ocr_post_df = pd.read_csv(" models-BLIP-hf/results/ocr-post/summary_report.csv")
ocr_post_desc_df = pd.read_csv(" models-BLIP-hf/results/ocr-post-desc/summary_report.csv")
post_ce_df = pd.read_csv(" models-BLIP-hf/results/post-wce/summary_report.csv")
blip_all_df = pd.read_csv(" models-BLIP-hf/results/all/summary_report.csv")

pix_df = pd.read_csv(" models-PIX-hf/results/base/summary_report.csv")
pix_576 = pd.read_csv(" models-PIX-hf/results/base-576/summary_report.csv")
pix_all =  pd.read_csv(" models-PIX-hf/results/all/summary_report.csv")
pix_wce =  pd.read_csv(" models-PIX-hf/results/wce/summary_report.csv")
pix_post = pd.read_csv(" models-PIX-hf/results/post/summary_report.csv")
git = pd.read_csv(" models-GIT-hf/results/base-384/summary_report.csv")
git_post_576 = pd.read_csv(" models-GIT-hf/results/post-576/summary_report.csv")
df = pd.concat([blip_base_df,ce_df,size_df,size_768_df ,desc_df,ocr_df,ocr_post_df,ocr_desc_df,ocr_post_desc_df,post_ce_df,blip_all_df,pix_df,pix_576,git,git_post_576])
df_pix = pd.concat([pix_df,pix_576,pix_wce,pix_post,pix_all])

df = df[df['average'].isin(['micro'])]
df_pix = df_pix[df_pix['average'].isin(['micro'])]
ic(df)
#sns.catplot(data=df, kind="bar", x="qtype", y="precision", hue="model")
```
